```python
import sox
import IPython.display as ipd

PATH = '/home/stjepan/guitar_data/'
file = PATH+"G53-41101-1111-00002.wav"
out = PATH+"guitar_out.wav"
```

### Defining different effects

```python

def overdrive(file, gain, colour):
    tfm = sox.Transformer()
    tfm.overdrive(gain_db=gain, colour=colour )
    tfm.vol(gain=0.3)
    tfm.build(file, out)
    return ipd.Audio(out)

def chorus(file, n_voices, delays, speeds, depths): 
    tfm = sox.Transformer()
    tfm.chorus(n_voices = n_voices, delays = delays, speeds = speeds, depths = depths)
    tfm.vol(gain=1)
    tfm.build(file, out)
    return ipd.Audio(out)

def phaser(file, gain_in, delay, decay,speed): 
    tfm = sox.Transformer()
    tfm.phaser( gain_in=gain_in, delay=delay, decay=decay, speed=speed)
    tfm.vol(gain=0.7)
    tfm.build(file, out)
    return ipd.Audio(out)

def reverb(file, reverberance, room_scale, wet_gain): 
    tfm = sox.Transformer()
    tfm.reverb(reverberance = reverberance, room_scale = room_scale, wet_gain=wet_gain)
    tfm.vol(gain=0.7)
    tfm.build(file, out)
    return ipd.Audio(out)

def echo(file, n_echos, delays, decays): 
    tfm = sox.Transformer()
    tfm.echo(n_echos = n_echos, delays = delays, decays = decays)
    tfm.vol(gain=0.7)
    tfm.build(file, out)
    return ipd.Audio(out)

```

### Testing overdrive

```python
#the best range is 15-80 for gain, and 0-100 for colour

overdrive(file, 45, 80)  
```
